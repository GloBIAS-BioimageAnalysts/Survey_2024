```python
import pandas
import utils

df = pandas.read_csv("../data/survey2024_headerscleaned.csv")
```

```python
short_colnames = {
    "bioimage analysis instruction": "instruction",
    "bioimage analysis service": "service",
    "bioimage analysis infrastructure development and maintenance": "infrastructure",
    "own bioimage analysis project": "own\nprojects",
    "bioimage analysis tool development not bound to a single service project": "tool\ndevelopment",
    "management or administration": "management/\nadministration",
}
long_colnames = {
    f"Q7 How much of your time do you spend on each of the following  [{k}]": v
    for k, v in short_colnames.items()
}

utils.normalized_percent_graphs(
    df,
    long_colnames,
    "../figures/WhereTimeSpent_filtered.png",
    prints=True,
    xlabel="Time Spent",
)
utils.normalized_percent_graphs(
    df,
    long_colnames,
    "../figures/WhereTimeSpent_filtered.svg",
    prints=True,
    xlabel="Time Spent",
)
```
