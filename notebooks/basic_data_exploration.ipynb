# Basic data exploration
This notebook serves as a initial approach to the data, so that we have an overview of the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/survey2024.csv")
```

```python
df.head()
```

```python
print (f"Total number of questions..: {df.shape[1]}")
print (f"Total number of answers....: {df.shape[0]}")
```

```python
for i, col in enumerate(df.columns):
    print(f"Column {i}")
    print(col)
    print(f"Data type: {df[col].dtype}")
    print(f"Answers: {(df[col].count() / df.shape[0]):.2%}")
    print("\n\n")
```
